```python
#grouping
#split - apply - combine 
import pandas as pd
df = pd.read_csv('https://raw.githubusercontent.com/dayanandv/Data-Science/main/dataset/weather_by_cities.csv')
df
```

```python
g = df.groupby('city')
g.first()  # gives first occurence 

```

```python
g.get_group('mumbai')
```

```python
g.max()
```

```python
g.mean('temperature')
```
